```python
import pandas as pd

# Load cleaned dataset
cleaned_path = "../data/AAPL_cleaned.csv"
df = pd.read_csv(cleaned_path)

# Check the first few rows
df.head()
```

```python

import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


# 0. Ensure root-level images folder exists

if not os.path.exists("../images"):  
    os.makedirs("../images")


# 1. Apple Stock Closing Price

plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.title('Apple Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.savefig("../images/closing_price.png", dpi=300)  # SAVE IMAGE
plt.show()

# 2. Apple Stock Price with Moving Averages

plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['MA7'], label='7-Day MA')
plt.plot(df['Date'], df['MA30'], label='30-Day MA')
plt.title('Apple Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.savefig("../images/moving_averages.png", dpi=300)  # SAVE IMAGE
plt.show()


# 3. Apple Stock Volatility

plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Volatility7'], label='7-Day Volatility')
plt.plot(df['Date'], df['Volatility30'], label='30-Day Volatility')
plt.title('Apple Stock Price Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.savefig("../images/volatility.png", dpi=300)  # SAVE IMAGE
plt.show()

#4.Interactive Plotly Closing Price

fig = px.line(df, x='Date', y='Close', title='Apple Closing Price Over Time')
fig.write_image("../images/plotly_closing_price.png")  # SAVE IMAGE (requires kaleido)
fig.show()
```
